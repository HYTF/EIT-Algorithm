```python
##RUN THIS CELL AND BELOW FIRST
#Import Packages
from datetime import datetime
import math
import copy
import numpy as np
import scipy
from scipy import io
from random import *
from scipy.linalg import svdvals
import h5py
import csv
import matplotlib.pyplot as plt
import pandas as pd
import random
from itertools import combinations
from itertools import combinations
import random
from numpy.random import randint
from numpy.random import rand
def import_csv(file):
    output = []
    with open(file) as csvfile:
        reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC) # change contents to floats
        for row in reader: # each row is a list
            #row.remove(row[-1])
            lst = list([])
            for h in row:
                lst.append(int(h))
            output.append(lst)
    return output
vector = []
#Volume Calculations

def volume_checksvd(basis):
    sensmx=[]
    for each in basis:
        sensmx.append(copy.deepcopy(sensitivity[charliemeasurements.index(each)]))
    svs=svdvals(sensmx)
    polyvol=np.prod(svs)
    return polyvol
def volume_check0(basis):
    d = 0
    exiting=0
    if d == 0:
        mu = sensitivity[charliemeasurements.index(basis[0])]
        mumag = np.linalg.norm(mu)
        muprimes = [mu]
        muprimesmag = [mumag]
        polyvol = mumag
        d += 1
    base = basis[1:]
    for each in base:
        mu2 = sensitivity[charliemeasurements.index(each)]
        mupar2 = np.zeros(sensitivity.shape[1],)
        for val in range(d):
            if muprimesmag[val]==0:
                exiting=1
                break
            mupar2 += (np.dot(muprimes[val], mu2)/(muprimesmag[val]**2)) * muprimes[val] 
        if exiting==1:
            break
        muprime = mu2 - mupar2
        muprimemag = np.linalg.norm(muprime)
        if muprimemag< np.linalg.norm(1e-14*np.ones_like(muprime)):
            muprimemag=0
        polyvol = muprimemag*polyvol
        muprimesmag.append(muprimemag)
        muprimes.append(np.array(muprime))
        d += 1
    return polyvol
def volume_check(basis):
    d = 0
    if d == 0:
        mu = sensitivity[charliemeasurements.index(basis[0])]
        mumag = np.linalg.norm(mu)
        muprimes = [mu]
        muprimesmag = [mumag]
        polyvol = mumag
        d += 1
    base = basis[1:]
    for each in base:
        mu2 = sensitivity[charliemeasurements.index(each)]
        mupar2 = np.zeros(sensitivity.shape[1],)
        for val in range(d):
            mupar2 += (np.dot(muprimes[val], mu2)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu2 - mupar2
        muprimemag = np.linalg.norm(muprime)
        polyvol = muprimemag*polyvol
        muprimesmag.append(muprimemag)
        muprimes.append(np.array(muprime))
        d += 1
    return polyvol

def volume_check2(basis):
    ##Note: The polyvol returned is not the volume of the polytope. This error does not effect the rest of the code. 
    d = 0
    if d == 0:
        mu = sensitivity[charliemeasurements.index(basis[0])]
        mumag = np.linalg.norm(mu)
        muprimes = [mu]
        muprimesmag = [mumag]
        polyvol = mumag
        d += 1
    base = basis[1:]
    for each in base:
        mu2 = sensitivity[charliemeasurements.index(each)]
        mupar2 = np.zeros(sensitivity.shape[1],)
        for val in range(d):
            mupar2 += (np.dot(muprimes[val], mu2)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu2 - mupar2
        muprimemag = np.linalg.norm(muprime)
        polyvol = muprimemag*polyvol
        muprimesmag.append(muprimemag)
        muprimes.append(np.array(muprime))
        d += 1
    return polyvol, muprimes, muprimesmag

#Perpendicular Components
def perpcomp(D2, basis, totalmeas):
    #gives a list of perpendicular components for each item in the basis compared to the rest of the basis. 
    #charliemeasurements is the complete list of measurements corresponding to the sensitivity matrix 
    perpcomplst = []
    mumags = []
    sineprimes = []
    for u in range(D2):
    #for 0, 1, ..., depth of basis - 1
        measurement = copy.deepcopy(basis)
        measurement.pop(u)
        #removes the one measurement you are looking at
        polyvol, muprimes, muprimesmag = volume_check2(measurement)
        #gets metrics on list
        thismes = copy.deepcopy(basis[u])
        #the measurement you are looking at
        d = D2- 1
        #length of the list with one removed
        mu = sensitivity[charliemeasurements.index(thismes)]
        mumag = np.linalg.norm(mu)
        #current mu
        mupar = np.zeros(sensitivity.shape[1],)
        #initialize mu parallel to array of zeros of the appropriate length
        for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu - mupar
        #this is your muprime when adding this measurement
        muprimemag = np.linalg.norm(muprime)
        #magnitude of mu prime
        sineprime = muprimemag/mumag
        perpcomplst.append(copy.deepcopy(muprimemag))
        mumags.append(copy.deepcopy(mumag))
        sineprimes.append(copy.deepcopy(sineprime))
        #appends magnitude of perp component to list        
    return perpcomplst, mumags, sineprimes

def singleperp(idx, basis, totalmeas):
    #gives the perpendicular component of one measurement in relation to the rest of the basis
    measurement = copy.deepcopy(basis)
    measurement.pop(idx)
    polyvol, muprimes, muprimesmag = volume_check2(measurement)
    thismes = copy.deepcopy(basis[idx])
    d = len(basis) - 1
    mu = sensitivity[charliemeasurements.index(thismes)]
    #mumag = np.linalg.norm(mu)
    mupar = np.zeros(sensitivity.shape[1],)
    for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
    muprime = mu - mupar
    muprimemag = np.linalg.norm(muprime)
    return muprimemag

def singleperp2(idx, basis, totalmeas):
    #same as singleperp, returns muprime as well as it's magnitude.
    measurement = copy.deepcopy(basis)
    measurement.pop(idx)
    polyvol, muprimes, muprimesmag = volume_check2(measurement)
    thismes = copy.deepcopy(basis[idx])
    d = len(basis) - 1
    mu = sensitivity[charliemeasurements.index(thismes)]
    #mumag = np.linalg.norm(mu)
    mupar = np.zeros(sensitivity.shape[1],)
    for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
    muprime = mu - mupar
    muprimemag = np.linalg.norm(muprime)
    return muprime, muprimemag



def mxsingleperp(trimmedbasis, totalmeas):
    #of the options available, picks the measurement that has the maximum perpenticular component value. 
    measurements = copy.deepcopy(trimmedbasis)
    polyvol, muprimes, muprimesmag = volume_check2(measurements)
    mxperp=0
    mxmes=[]
    for option in totalmeas:
        if (option not in trimmedbasis):
            thismes = copy.deepcopy(option)
            d = len(trimmedbasis) 
            mu = sensitivity[charliemeasurements.index(thismes)]
    #mumag = np.linalg.norm(mu)
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            muprime = mu - mupar
            muprimemag = np.linalg.norm(muprime)
            print(muprimemag)
            if muprimemag>mxperp:
                mxperp=copy.deepcopy(muprimemag)
                print(mxperp)
                mxmes=copy.deepcopy(option)
    return mxperp, mxmes
def mxsingletheta(trimmedbasis, totalmeas):   
    #of the options available, picks the measurement that has the maximum sin2theta 
    measurements = copy.deepcopy(trimmedbasis)
    polyvol, muprimes, muprimesmag = volume_check2(measurements)
    mxtheta=0
    mxmes=[]
    for option in totalmeas:
        if (option not in trimmedbasis):
            thismes = copy.deepcopy(option)
            d = len(trimmedbasis) 
            mu = sensitivity[charliemeasurements.index(thismes)]
    #mumag = np.linalg.norm(mu)
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            mumag=np.linalg.norm(mu)
            muprime = mu - mupar
            muprimemag=np.linalg.norm(muprime)
            sin2theta=(muprimemag**2)/(mumag**2)
            if sin2theta>mxtheta:
                mxtheta=copy.deepcopy(sin2theta)
                mxmes=copy.deepcopy(option)
    return mxtheta, mxmes
def perpadd(D,basis,totalmeas):
    #returns a data frame of measurements and their perp components wrt others for each subbasis
    for u in range(D+1):
        if u>1:
            subbasis=[basis[i] for i in range(u)]
            D2=len(subbasis)
            perpsteps, mumagsteps, sineprimesteps=perpcomp(D2,subbasis,totalmeas)
            dataframeperp = {'measurements': subbasis, 'perpendicular component wrt others': perpsteps, 'vector magnitudes': mumagsteps, 'sines': sineprimesteps}
            dfp = pd.DataFrame(dataframeperp, columns = ['measurements', 'perpendicular component wrt others', 'vector magnitudes', 'sines'])
            #dfp.to_csv('singel8finer.csv', mode='a')
            dataframeperp2 = {'measurements': subbasis, 'perpendicular component wrt others': perpsteps}
            dfp2 = pd.DataFrame(dataframeperp2, columns = ['measurements', 'perpendicular component wrt others'])
            #print(dfp2)
    return dfp2

def pertheta(D,basis,totalmeas):
        #commented out: returns a data frame of measurements and their perp components wrt others for each subbasis
    #for u in range(D+1):
        #if u>1:
            #subbasis=[basis[i] for i in range(u)]
            #D2=len(subbasis)
            subbasis=copy.deepcopy(basis)
            D2=len(subbasis)
            perpsteps, mumagsteps, sineprimesteps=perpcomp(D2,subbasis,totalmeas)
            svals=singvals(basis,sensitivity, charliemeasurements)
            svs=svals.tolist()
            dataframeperp = {'measurements': subbasis, 'perpendicular component wrt others': perpsteps, 'vector magnitudes': mumagsteps, 'sines': sineprimesteps}
            dfp = pd.DataFrame(dataframeperp, columns = ['measurements', 'perpendicular component wrt others', 'vector magnitudes', 'sines'])
            #dfp.to_csv('singel8finer.csv', mode='a')
            dataframeperp2 = {'measurements': subbasis, 'sine theta': sineprimesteps, 'MOP values': perpsteps, 'vector magnitudes': mumagsteps}
            dfp2 = pd.DataFrame(dataframeperp2, columns = ['measurements', 'sine theta', 'MOP values', 'vector magnitudes'])
            #print(dfp2)
            return dfp2, svs
    #return dfp2

def orderedpertheta_plusV(sze,basis,totalmeas):
    #prints dataframe of perp comp wrt others in decending order. Returns B''/A'' (smallest/largest perp comp)
    vol=volume_check(basis)
    print("Volume: ")
    print(vol)
    df,svs=pertheta(sze,basis,totalmeas)
    sorted_df = df.sort_values(by=['sine theta'], ascending=False)
    sinethetas = sorted_df['sine theta'].tolist()
    sorted_df.insert(4,'singular vals', svs, True)
    # print("A''- B'' =  ")
    # print(mocomps[0]-mocomps[-1])
    # print("B''/A''")
    # print(mocomps[-1]/mocomps[0])
    # print("sigma_M'")
    #print(sigmamp(mocomps,0.001))
    #print(simply_sigma(basis))
    print(sorted_df)
    return min(sinethetas)
   # return mocomps[0] - mocomps[-1] 
    #return sorted_df


def orderedperpadd_plusV(D,basis,totalmeas):
    #prints dataframe of perp comp wrt others in decending order. Returns B''/A'' (smallest/largest perp comp)
    vol=volume_check(basis)
    print("Volume: ")
    print(vol)
    df=perpadd(D,basis,totalmeas)
    sorted_df = df.sort_values(by=['perpendicular component wrt others'], ascending=False)
    mocomps = sorted_df['perpendicular component wrt others'].tolist()
    print("A''- B'' =  ")
    print(mocomps[0]-mocomps[-1])
    print("B''/A''")
    print(mocomps[-1]/mocomps[0])
    print("sigma_M'")
    #print(sigmamp(mocomps,0.001))
    print(simply_sigma(basis))
    print(sorted_df)
    return mocomps[0] - mocomps[-1] 
    #return sorted_df

def perpadd2(D,basis,totalmeas):
    #returns list of perp components wrt others in decending order and a list of ordered measurements
    df=perpadd(D,basis,totalmeas)
    sorted_df = df.sort_values(by=['perpendicular component wrt others'], ascending=False)
    mocomps = sorted_df['perpendicular component wrt others'].tolist()
    orderedmes = sorted_df['measurements'].tolist()
    return orderedmes, mocomps

def next_vector(d, vector, totalmeas, measurements, polyvol, muprimesmag, muprimes, sinthetas):  
    #finds next addition to maximize sin value of top 2% of volumes
    maxvol = 0
    maxvec= []
    maxmag = 0
    muprime = []
    maxmuprime = []
        
    volumes = []
    sinthetasquared = []
    newvol = []
    newmes = []
    
    for each2 in measurements:
        #print(each2)
        test_vector = copy.deepcopy(vector)
        test_vector.append(each2)
        mu = sensitivity[charliemeasurements.index(each2)]
        mupar = np.zeros(sensitivity.shape[1],)
        for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu - mupar
        muprimemag = np.linalg.norm(muprime)
        answer = muprimemag*polyvol
        volumes.append(copy.deepcopy(answer))
        sinthetasquared.append((copy.deepcopy(muprimemag))**2 / (np.linalg.norm(copy.deepcopy(mu)))**2)
        if answer >= maxvol:
            maxvol = copy.deepcopy(answer)

        
    window = maxvol * .02

    for j in volumes:
        if j >= maxvol - window:
            if j <= maxvol + window:
                newvol.append(j)
                newmes.append(measurements[volumes.index(j)])
        
    maxmax = 0
    maxmes = []
    for h in newmes:
        i = sinthetasquared[measurements.index(h)]
        if i >= maxmax:
            maxmax = i
            maxmes = h
        
        
    test_vector = copy.deepcopy(vector)
    test_vector.append(maxmes)
    mu = sensitivity[charliemeasurements.index(maxmes)]
    mupar = np.zeros(sensitivity.shape[1],)
    for val in range(d):
        mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
    muprime = mu - mupar
    muprimemag = np.linalg.norm(muprime)
    answer = muprimemag*polyvol
        
    maxvol = copy.deepcopy(answer)
    maxmag = copy.deepcopy(muprimemag)
    maxvec = copy.deepcopy(test_vector)
    maxmuprime = copy.deepcopy(muprime)
        
    new_measurements = copy.deepcopy(measurements)
    new_measurements.pop(measurements.index(maxvec[-1]))
    new_muprimes = copy.deepcopy(muprimes)
    new_muprimesmag = copy.deepcopy(muprimesmag)
    new_muprimesmag.append(maxmag)
    new_muprimes.append(np.array(maxmuprime))
    sinthetas.append(copy.deepcopy(maxmax))
        
    return maxvec, maxvol, sinthetas, new_muprimesmag, new_muprimes, new_measurements


def singleCO(d, D, vector, totalmeas, measurements, polyvol, muprimesmag, muprimes, sinthetas,firstroundmaxmes):
    #does a single run with the maximum measurement from first round
    mags = []

    
    if d == 0:
        test_measurements = copy.deepcopy(measurements)
        test_measurements.pop(measurements.index(firstroundmaxmes))
        mu = sensitivity[charliemeasurements.index(firstroundmaxmes)]
        mumag = np.linalg.norm(mu)
        
        basis, vol, thetas, magmuprimes, muprimes = max_polytopeCO(d+1, D, [firstroundmaxmes], totalmeas, test_measurements, mumag, [mumag], [mu], [0])
        
        for b in basis:
            mags.append(np.linalg.norm(sensitivity[charliemeasurements.index(b)]))
        
        
        X = []
        for j in range(D):
            X.append(j + 1)
        
        
        
        return basis, vol, thetas, magmuprimes, muprimes

def magchop(choppct):
    choppedmtx=[]
    choppedmeas=[]
    vecmags=[]
    for vector in sensitivity:
        vecmag=np.linalg.norm(vector)
        vecmags.append(vecmag)
    sortedmags=copy.deepcopy(vecmags)
    sortedmags.sort(reverse=True)
    #print(sortedmags[0])
    dim=len(sensitivity)
    chopdec=choppct/100;
    pct=round(dim*chopdec);
    #print(pct-1)
    if choppct>0:
        chop=copy.deepcopy(sortedmags[pct-1])
    elif choppct==0:
        chop=1+copy.deepcopy(sortedmags[0])
    print(chop)
    sensmx=copy.deepcopy(sensitivity)
    for msi,u in enumerate(sensmx):
        v=np.linalg.norm(u)
        if v<chop:
            choppedmtx.append(u)
            ms=copy.deepcopy(charliemeasurements[msi])
            choppedmeas.append(ms)
    return choppedmtx, choppedmeas

def singvals(basis,smatrix,measlst):
    measurements=copy.deepcopy(basis)
    mx4svd=[]
    for u in measurements:
        sensrow=copy.deepcopy(smatrix[measlst.index(u)])
        mx4svd.append(sensrow)
    U,SD,VT=np.linalg.svd(mx4svd)
    return SD



def guidedgreedy2(D, vector, totalmeas, measurements, polyvol, muprimesmag, muprimes, sinthetas):
    start_time = datetime.now()
    #same as guidedgreedy but compares polyvol instead of MOPs, appears to give same answers 
    firstroundmax = 0
    firstroundmaxmes = []
    
    for each in measurements:
        mu = sensitivity[charliemeasurements.index(each)]
        mumag = np.linalg.norm(mu)
        if mumag >= firstroundmax:
            firstroundmax = mumag
            firstroundmaxmes = each
            
    test_measurements = copy.deepcopy(measurements)
    test_measurements.pop(measurements.index(firstroundmaxmes))
    mu = sensitivity[charliemeasurements.index(firstroundmaxmes)]
    mumag = firstroundmax
    stp=0
    while stp < D-1:
        if stp==0:
            maxvec, maxvol, sinthetas, new_muprimesmag, new_muprimes, new_measurements = next_vector(stp+1, [firstroundmaxmes], totalmeas, test_measurements, mumag, [mumag], [mu], [0])
            #perpsteps, mumagsteps, sineprimesteps=perpcomp(stp+2,maxvec,charliemeasurements)
            #dataframeperp2 = {'measurements': maxvec, 'perpendicular component wrt others': perpsteps}
            #dfp2 = pd.DataFrame(dataframeperp2, columns = ['measurements', 'perpendicular component wrt others'])
            #print(dfp2)
            stp=stp+1
            #print("Basis Build: ")
            #print(maxvec)
        else:
            maxvec, maxvol, sinthetas, new_muprimesmag, new_muprimes, new_measurements = next_vector(stp+1, maxvec, totalmeas, new_measurements, maxvol, new_muprimesmag, new_muprimes, sinthetas)
            #perpsteps, mumagsteps, sineprimesteps=perpcomp(stp+2,maxvec,charliemeasurements)
            stp=stp+1
            # print("Basis Build: ")
            # print(maxvec)
            #dataframeperp2 = {'measurements': maxvec, 'perpendicular component wrt others': perpsteps}
            #dfp2 = pd.DataFrame(dataframeperp2, columns = ['measurements', 'perpendicular component wrt others'])
            #print(dfp2)  
            #print(added)
            #first=maxvec[0]
            #print(first)
            newbasis=guide2(maxvec,totalmeas)
            #print(len(newbasis))
            perpsteps2=perpcomp(len(newbasis),newbasis,totalmeas)
            largest=max(perpsteps2)
            lidx=perpsteps2.index(largest)
            lg=newbasis[lidx]
            # print(lg)
            # print(len(lg))
            maxvec, maxvol, sinthetas, new_muprimesmag, new_muprimes=singleCO(0, len(newbasis), vector, totalmeas, newbasis, 0, [], [], [],lg)
    print(maxvec)
    end_time = datetime.now()
    print('Duration: {}'.format(end_time - start_time))
    #print(new_measurements)
    return maxvec

def mxvolchecklim(trimmedbasis, allmeas, fract):
    #start_time=datetime.now()
    #of the options available, picks the measurement that maximizes the volume 
    measurements = copy.deepcopy(trimmedbasis)
    polyvol, muprimes, muprimesmag = volume_check2(measurements)
    mxvol=0
    totalmeas=random.sample(allmeas,int(np.floor(fract*len(allmeas))))
    trialbs=copy.deepcopy(trimmedbasis)
    for option in totalmeas:
        if (option not in trimmedbasis):
            thismes = copy.deepcopy(option)
            trialbs.append(thismes)
            #vol=volume_check(trialbs)
            d = len(trimmedbasis) 
            mu = sensitivity[allmeas.index(thismes)]
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            muprime = mu - mupar
            muprimemag = np.linalg.norm(muprime)
            vol=muprimemag*polyvol
            if vol>mxvol:
                mxvol=copy.deepcopy(vol)
                mxmes=copy.deepcopy(option)
    end_time = datetime.now()
    #print('Duration: {}'.format(end_time - start_time))
    return mxvol, mxmes     

# def mxvolchecksvd(trimmedbasis, totalmeas):
#     #of the options available, picks the measurement that maximizes the volume 
#     measurements = copy.deepcopy(trimmedbasis)
#     #polyvol = volume_checksvd(measurements)
#     mxvol=0
#     mumx=[]
#     trialbs=copy.deepcopy(trimmedbasis)
#     for option in totalmeas:
#         if (option not in trimmedbasis):
#             thismes = copy.deepcopy(option)
#             trialbs.append(thismes)
#             #vol=volume_check(trialbs)
#             d = len(trimmedbasis) 
#             mumx.append(copy.deepcopy(sensitivity[charliemeasurements.index(thismes)]))
#             #mupar = np.zeros(sensitivity.shape[1],)
#             #for val in range(d):
#                 #mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
#             #muprime = mu - mupar
#             muprimemag = np.linalg.norm(muprime)
            
#             if vol>mxvol:
#                 mxvol=copy.deepcopy(vol)
#                 mxmes=copy.deepcopy(option)
#     return mxvol, mxmes 

def mxvolcheck(trimmedbasis, totalmeas):
    #of the options available, picks the measurement that maximizes the volume 
    measurements = copy.deepcopy(trimmedbasis)
    polyvol, muprimes, muprimesmag = volume_check2(measurements)
    mxvol=0
    trialbs=copy.deepcopy(trimmedbasis)
    for option in totalmeas:
        if (option not in trimmedbasis):
            thismes = copy.deepcopy(option)
            trialbs.append(thismes)
            #vol=volume_check(trialbs)
            d = len(trimmedbasis) 
            mu = sensitivity[charliemeasurements.index(thismes)]
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            muprime = mu - mupar
            muprimemag = np.linalg.norm(muprime)
            vol=muprimemag*polyvol
            if vol>mxvol:
                mxvol=copy.deepcopy(vol)
                mxmes=copy.deepcopy(option)
    return mxvol, mxmes         
        
def guide2(maxvec,totalmeas): #new_measurements
    #checks for maximized basis by comparing if a new addition of one of the other measurements has a better perp component than the one added
    added=copy.deepcopy(maxvec[-1])
    #new_measurements.pop(new_measurements.index(added))
    checking=copy.deepcopy(maxvec)
    checking.pop(maxvec.index(added))
    new=copy.deepcopy(maxvec)
    # print("NEW(OG):")
    # print(new)
    # perpsteps, mumagsteps, sineprimesteps=perpcomp(len(new),new,charliemeasurements)
    # dataframeperp2 = {'measurements': maxvec, 'perpendicular component wrt others': perpsteps}
    # dfp2 = pd.DataFrame(dataframeperp2, columns = ['measurements', 'perpendicular component wrt others'])
    #print(dfp2)
    checked=[]
    checked.append(copy.deepcopy(added))
    # print("Checking: ")
    # print(checking)
    for each in checking:
        # print("Each: ")
        # print(each)
        testbasis=copy.deepcopy(new)
        testbasis.pop(testbasis.index(each))
        #ogperp=singleperp(new.index(each),new,charliemeasurements)
        #oldperp=copy.deepcopy(ogperp)
        
        mxvol,mxmes=mxvolcheck(testbasis, totalmeas)
        #print(singvals(testbasis,sensitivity,charliemeasurements))
        #checking.pop(checking.index(each))
        #checking.append(added)
        if each==mxmes:
            #print('NO CHANGE')
            checked.append(copy.deepcopy(each))
        else:
            #print('NEW MEAS')
            #checked.pop(checked.index(added))
            for c in checked:
                checking.append(copy.deepcopy(c))
            #checking.append(copy.deepcopy(each))
            new.pop(new.index(each))
            new.append(mxmes)
            checked=[]
            checked.append(copy.deepcopy(mxmes))       
    # print("NEW:")
    # print(new)
    return new

def MC_singvals(basis,totalmeas,sensmx,count,beta,svlim):
    d=len(basis)
    bsis=copy.deepcopy(basis)
    meassea=[]
    # vecsea=[]
    # keptmx=[]
    kept=[]
    ocean=copy.deepcopy(totalmeas)
    for uidx,u in enumerate(ocean):
        if u in bsis:
            kept.append(u)
        #     keptmx.append(sensitivity[uidx])
        if u not in kept:
            meassea.append(u)
            # vecsea.append(sensitivity[uidx])
    svs=singvals(kept,sensmx,totalmeas)
    mx=max(svs)
    mn=min(svs)
    V=np.prod(svs)
    AB=(mx-mn)**d
    gmma_mx=(1-beta)*V-(beta)*AB
    print(gmma_mx)
    #Now we have a start to compare with 
    counter=0
    while counter<count:
        trialtoss=copy.deepcopy(kept)
        trial=copy.deepcopy(kept)
        toss=random.sample(trial,1)
        toss1=toss[0]
        tsix=trialtoss.index(toss1)
        trial.pop(tsix)
        grab=random.sample(meassea,1)
        grab1=grab[0]
        trial.append(grab1)
        bix=meassea.index(grab1)
        gbix=copy.deepcopy(bix)
        svs_test=singvals(trial,sensmx,totalmeas)
        tmx=max(svs_test)
        tmn=min(svs_test)
        Vt=np.prod(svs_test)
        ABt=(tmx-tmn)**d
        #print(AB)
        gmma_trial=(1-beta)*Vt-(beta)*ABt
    
        if (gmma_trial>gmma_mx) and tmn>svlim:
            gmma_mx=copy.deepcopy(gmma_trial)
            kept=copy.deepcopy(trial)
            counter=0
            meassea.append(copy.deepcopy(toss1))
            meassea.pop(gbix)
            print("NEW")
            print(tmx-tmn)
            print(V)
            print(gmma_trial)
        else:
            counter=counter+1
            #print("No Change")
            #print(counter)
    return kept
def svpairs(d):
    first=d-1
    if f<10:
        many=1+math.ceil(f/2)
    #else
#U,SD,VT=np.linalg.svd(G)
#V=np.transpose(VT)
#UT=np.transpose(U)


def planarproj(fitvecs,sensvecs):
    sensvectors=copy.deepcopy(sensvecs)
    # normal=np.cross(fitveca,fitvecb)
    # nhat=copy.deepcopy(normal)
    #nhat=normal/np.linalg.norm(normal)
    projections=[]
    for u in sensvectors:
        # pdenom=np.dot(nhat,nhat)
        # pnum=np.dot(u,nhat)
        # pfrac=pnum/pdenom
        # ppar=pfrac*nhat
        # proje=u-ppar
        proje=singleperp3(u,fitvecs)
        projections.append(proje)
    return projections

def singleperp3(meas, planevectors):
    #same as singleperp but uses external vectors in model space
    thismes = copy.deepcopy(meas)
    muprimes, muprimesmag = planemake(planevectors)
    d = len(planevectors)
    mu = copy.deepcopy(thismes)
    #mumag = np.linalg.norm(mu)
    mupar = np.zeros(mu.shape,)
    for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
    muprime = mu - mupar
    muprimemag = np.linalg.norm(muprime)
    return mupar

def planemake(planebuild):
    planevecs=copy.deepcopy(planebuild)
    d = 0
    if d == 0:
        mu = copy.deepcopy(planevecs[0])
        mumag = np.linalg.norm(mu)
        muprimes = [mu]
        muprimesmag = [mumag]
        #polyvol = mumag
        d += 1
    plne = planevecs[1:]
    for each in plne:
        #print(each)
        mu2 = each
        mupar2 = np.zeros(each.shape,)
        for val in range(d):
            mupar2 += (np.dot(muprimes[val], mu2)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu2 - mupar2
        muprimemag = np.linalg.norm(muprime)
        #polyvol = muprimemag*polyvol
        muprimesmag.append(muprimemag)
        muprimes.append(np.array(muprime))
        d += 1
    return muprimes, muprimesmag
def greedypair(D,fitvecs,sensmx,measures):
    projections=planarproj(fitvecs,sensmx)
    sensvecs=copy.deepcopy(projections)
    firstroundmax = 0
    firstroundmaxmes = []
    measurements=copy.deepcopy(measures)
    for each in measurements:
        mu = sensvecs[measurements.index(each)]
        mumag = np.linalg.norm(mu)
        measurements2=copy.deepcopy(measurements)
        measurements2.pop(measurements.index(each))
        for each2 in measurements2:
            mu2=sensvecs[measurements2.index(each2)]
            mu2mag = np.linalg.norm(mu2)
            tbs=[]
            tbs.append(each)
            tbs.append(each2)
            tbsvecs=[]
            tbsvecs.append(mu)
            tbsvecs.append(mu2)
            tvol=volume_check_external(tbs,projections,measures)
            
            if tvol >= firstroundmax:
                firstroundmax = tvol
                firstroundmaxmes = tbs
    print(firstroundmaxmes)
    end_time = datetime.now()
    print('Duration: {}'.format(end_time - start_time))
    #print(new_measurements)
    return firstroundmax
def volume_check_external(basis,smx,mlst):
    d = 0
    if d == 0:
        mu = smx[mlst.index(basis[0])]
        mumag = np.linalg.norm(mu)
        muprimes = [mu]
        muprimesmag = [mumag]
        polyvol = mumag
        d += 1
    base = basis[1:]
    for each in base:
        mu2 = smx[mlst.index(each)]
        mupar2 = np.zeros(mu2.shape,)
        for val in range(d):
            mupar2 += (np.dot(muprimes[val], mu2)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu2 - mupar2
        muprimemag = np.linalg.norm(muprime)
        polyvol = muprimemag*polyvol
        muprimesmag.append(muprimemag)
        muprimes.append(np.array(muprime))
        d += 1
    return polyvol


#sensitivity=np.array(planarproj(planevecs,Jacobian))

def guideloop(d,proto):
    #proto=guidedgreedy2(d,vector,charliemeasurements,charliemeasurements,[],[],[],[])
    mpto=[]
    tbs=copy.deepcopy(proto)
    same=0
    wcount=0
    while same==0 and wcount<d:
        #print(volume_check0(tbs))
        mpto=copy.deepcopy(guide2(tbs,charliemeasurements))
        if mpto==tbs:
            same=same+1
        else:
            if volume_check0(mpto)>=volume_check0(tbs):
                #print('better')
                tbs=copy.deepcopy(mpto)
            else:
                print('worse')
                wcount=wcount+1
                tbs=copy.deepcopy(proto)
                #print(volume_check(copy.deepcopy(tbs)))
                #print(tbs)
                tbs.append(copy.deepcopy(tbs[0]))
                tbs.pop(0)
                #print(volume_check0(tbs))
                #print(np.shape(tbs))
                
            #tbs=copy.deepcopy(mpto)
    print(volume_check(mpto))
    return mpto

def guideloop2(d,proto):
    #proto=guidedgreedy2(d,vector,charliemeasurements,charliemeasurements,[],[],[],[])
    mpto=[]
    tbs=copy.deepcopy(proto)
    same=0
    wcount=0
    while same==0 and wcount<d:
        #print(volume_check0(tbs))
        mpto=copy.deepcopy(guide2(tbs,charliemeasurements))
        if mpto==tbs:
            same=same+1
        else:
            if volume_check0(mpto)>=volume_check0(tbs):
                #print('better')
                tbs=copy.deepcopy(mpto)
            else:
                print('worse')
                wcount=wcount+1
                tbs=copy.deepcopy(proto)
                #print(volume_check(copy.deepcopy(tbs)))
                #print(tbs)
                tbs.append(copy.deepcopy(tbs[0]))
                tbs.pop(0)
                #print(volume_check0(tbs))
                #print(np.shape(tbs))
                
            #tbs=copy.deepcopy(mpto)
    print(volume_check(mpto))
    return mpto
def max_polytopeCO(d, D, vector, totalmeas, measurements, polyvol, muprimesmag, muprimes, sinthetas):
    #same as my max_polytope function
    if d == D:
        return vector, polyvol, sinthetas, muprimesmag, muprimes
    else:
        maxvol = 0
        maxvec= []
        maxmag = 0
        muprime = []
        maxmuprime = []
        
        volumes = []
        sinthetasquared = []
        newvol = []
        newmes = []
        
        for each2 in measurements:
            test_vector = copy.deepcopy(vector)
            test_vector.append(each2)
            mu = sensitivity[charliemeasurements.index(each2)]
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            muprime = mu - mupar
            muprimemag = np.linalg.norm(muprime)
            answer = muprimemag*polyvol
            volumes.append(copy.deepcopy(answer))
            sinthetasquared.append((copy.deepcopy(muprimemag))**2 / (np.linalg.norm(copy.deepcopy(mu)))**2)
            if answer >= maxvol:
                maxvol = copy.deepcopy(answer)

        
        window = maxvol * .02

        for j in volumes:
            if j >= maxvol - window:
                if j <= maxvol + window:
                    newvol.append(j)
                    newmes.append(measurements[volumes.index(j)])
        
        maxmax = 0
        maxmes = []
        for h in newmes:
            i = sinthetasquared[measurements.index(h)]
            if i >= maxmax:
                maxmax = i
                maxmes = h
        
        
        test_vector = copy.deepcopy(vector)
        test_vector.append(maxmes)
        mu = sensitivity[charliemeasurements.index(maxmes)]
        mupar = np.zeros(sensitivity.shape[1],)
        for val in range(d):
            mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
        muprime = mu - mupar
        muprimemag = np.linalg.norm(muprime)
        answer = muprimemag*polyvol
        
        maxvol = copy.deepcopy(answer)
        maxmag = copy.deepcopy(muprimemag)
        maxvec = copy.deepcopy(test_vector)
        maxmuprime = copy.deepcopy(muprime)
        
        new_measurements = copy.deepcopy(measurements)
        new_measurements.pop(measurements.index(maxvec[-1]))
        new_muprimes = copy.deepcopy(muprimes)
        new_muprimesmag = copy.deepcopy(muprimesmag)
        new_muprimesmag.append(maxmag)
        new_muprimes.append(np.array(maxmuprime))
        sinthetas.append(copy.deepcopy(maxmax))
        
        return max_polytopeCO(d+1, D, maxvec, totalmeas, new_measurements, maxvol, new_muprimesmag, new_muprimes, sinthetas)
    
    #!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun May 15 16:39:32 2022

@author: claireonsager
"""
#https://machinelearningmastery.com/simple-genetic-algorithm-from-scratch-in-python/ ****HEAVILY USED
#https://algodaily.com/lessons/introduction-to-genetic-algorithms-in-python
#https://www.geeksforgeeks.org/genetic-algorithms/
#https://blog.paperspace.com/genetic-algorithm-applications-using-pygad/
#Select functions modified from the above links' codes
#%%
#WARNING!!!! IF n_pop<<n_bits ONLY n_mating=1 WORKS!!!
#objective function: 
def objective(chrom):
    basis=[]
    vectors=[]
    for gene in chrom:
        basis.append(copy.deepcopy(charliemeasurements[gene]))
    #print(basis)
    if len(set(chrom)) < len(chrom):
        V=0
    else:
        V=volume_check0(basis)
    return V

def objectivesvd(chrom, sensitivity):
    sensmx=[]
    #print('newit')
    for gene in chrom:
        #print(gene)
        sensmx.append(1*copy.deepcopy(sensitivity[gene]))
    svs=svdvals(sensmx)
    #print(svs)
    if len(svs)<len(chrom):
        V=0
    if np.any(svs<1e-17):
        V=0
    else:
        V=np.prod(svs)
    #print('Vol')
    #print(V)
    return V

#Selecting best parents (Fitness proportionate selection)
def parentalweigh(fitness): #fitness is reordered to be 
    pwl=0
    pwu=0
    sumf=sum(fitness)
    weights=[]
    weightsu=[]
    weightsl=[]
    wN=[] #fitness/sum(fitness)
    #for c in fitness:
    for c in range(len(fitness)): #c is chrom index
        w=fitness[c]/sumf
        wN.append(w)
        weights.append(w)
        if c==0: #The first fitness, which is the lowest
            pwl=0
        else:
            pwl=pwl+wN[c-1]
        pwu=pwu+w
        weightsl.append(pwl) #lower bound for fitness proportional selection
        weightsu.append(pwu) #upper bound for fitness proportional selection
    return weights,weightsl,weightsu
def worst2(weights):
    worst=[]
    neww=copy.deepcopy(weights)
    mnf=np.where(neww==min(neww))
    mn=mnf[0][0]
    worst.append(mn)
    neww[mn]=1
    mnf2=np.where(neww==min(neww))
    mn2=mnf2[0][0]
    worst.append(mn2)
    return worst
def selectparents(pop,fitness,n_mating,n_bits):
    weights,weightsl,weightsu=parentalweigh(fitness)
    parents=[]
    # print('here')
    for po in range(n_mating):
        selectparam=random.uniform(0, 1)
        #selectparam=random.sample(range(0, 1), 2)
        foundparent=False
        while foundparent==False:
            for sp in range(len(weightsl)):
                if (selectparam>= weightsl[sp]) and (selectparam<weightsu[sp]):
                    parent1=sp
                    foundparent=True
        #print(parent1)
        parents.append(pop[parent1])
        #selectparam=random.sample(range(0, 1), 2)
        triesfor2=0
        foundparent2=False
        # while foundparent2==False:
        #     secondparentpop=copy.deepcopy(pop)
        #     secondparentfit=copy.deepcopy(fitness)
        #     secondparentpop.pop(parent1)
        #     secondparentfit.pop(parent1)
        #     weights2,weightsl2,weightsu2=parentalweigh(secondparentfit)
        #     selectparam2=random.uniform(0, 1)
        #     for sp2 in range(len(weightsl2)):
        #         if (selectparam2>= weightsl2[sp2]) and (selectparam2<weightsu2[sp2]):
        #             parent2b=sp2
        #             foundparent2=True
        # parents.append(secondparentpop[parent2b])
        # parent2=np.where(pop==secondparentpop[parent2b])
        while foundparent2==False:
            triesfor2=triesfor2+1
            #print(triesfor2)
            selectparam2=random.uniform(0, 1)
            for sp in range(len(weightsl)):
                if (selectparam2>= weightsl[sp]) and (selectparam2<weightsu[sp]):
                    parent2=sp
                    #print(parent2)
                    if parent2!=parent1:
                        foundparent2=True
                    elif triesfor2>1000:
                        parent2=-1
                        foundparent2=True
        if parent2==-1:
            parents.append(random.sample(range(len(sensitivity)),n_bits))
        else:
            parents.append(pop[parent2])
    return parents,weights
    
def mating(pop,parents,rcross,rmut,n_bits,n_mating):
    #single point crossover now for convenience
    offspringSize=2*n_mating
    #offspringSize=2
    offspring = np.empty((offspringSize, n_bits))
    #print(offspring)
    #print(parents)
    offspring[0]=parents[0]#pop[parents[0]]
    offspring[1]=parents[1]#pop[parents[1]]
    #print("offspring")
    #print(offspring)
    #Crossover
    #print('now')
    for kk in range(offspringSize): 
        # check for recombination
       # print('now')
        if rand() < rcross:
            #print('now')
            crossoverPoint=np.random.randint(0,n_bits)
            # Extracting first half of the offspring
            offspring[kk, 0: crossoverPoint] = parents[0][0:crossoverPoint]#pop[parents[0]][0:crossoverPoint]
            # Extracting second half of the offspring
            offspring[kk, crossoverPoint:] = parents[1][crossoverPoint:]
    #print(offspring)
    newoffspring=[]
    for o in offspring:
        newo=copy.deepcopy(o)
        for i in range(len(o)):
		# check for a mutation
        	if rand() < rmut:
                 #print('here')
			     # flip the bit
                 r= random.randrange(len(sensitivity))
                 while r in newo:
                     r= random.randrange(len(sensitivity))
                 newo[i]=r
        newoffspring.append(newo)
        #print(newoffspring)
    return newoffspring

# def oneptcross:
    
# def multiptcross:

# def uniformcross:
#global best
def genetic(n_bits, n_iter, n_pop, rcross, rmut,n_mating,n_greed,n_swaps,greedlim, sens, charliemeasurements):#(n_bits, n_iter, n_pop, n_parents, rcross, rmut):
    start_time = datetime.now()
    global sensitivity
    sensitivity = sens
    D = n_bits
    #define initial population
    pop = [ random.sample(range(len(sensitivity)),n_bits) for _ in range(n_pop)]
    #print(pop)
    # keep track of best solution
    initialrmut=copy.deepcopy(rmut)
    initialrcross=copy.deepcopy(rcross)
    global best
    best = pop[0]
    best_eval=objectivesvd(pop[0], sensitivity, D)
    uselessgens=0
    # enumerate generations
    for gen in range(n_iter):
        rmut=copy.deepcopy(initialrmut)
        if gen>0.75*n_iter:
            rmut=0.3#math.floor(initialrmut/2)
            rcross=0.7#math.ceil(initialrcross*2)
            if uselessgens>=51:
                uselessgens=0
        #print(gen)
        # evaluate all candidates in the population
        #print(pop)
        if gen==0:
            fitness = [objectivesvd(c, sensitivity, D) for c in pop]
        pop = [x for _,x in sorted(zip(fitness,pop))]
        #print('passed')
       # pop=np.array(pop)
        fitness = [x for x in sorted(fitness)]
        #print(fitness)
        # check for new best solution
        #for p in range(n_pop):
            #print('here')
        # print(fitness[0])
        # print(fitness[-1])
        if fitness[-1] > best_eval:
                #print('here2')
            best, best_eval = pop[-1], fitness[-1]
            #print("gen:%d, new best: %s) = %E" % (gen,  pop[-1], fitness[-1]))
            elapsed_time = datetime.now()-start_time
            print("gen:%d, duration:%s, new best = %E" % (gen, elapsed_time, fitness[-1]))
            uselessgens=0
            rmut=copy.deepcopy(initialrmut)
        else:
            uselessgens=uselessgens+1
        if uselessgens==50:
            #print("update rmut")
            rmut=1
            #uselessgens=0
        parents,weights=selectparents(pop,fitness,n_mating,n_bits)
        if uselessgens==n_greed:
            print('greed')
            #print(children[0])
            betterbasis=genguide(copy.deepcopy(best),charliemeasurements,n_swaps,n_bits,greedlim)            
            #print(children[0])
            #print(objective(children[0]))   
        #parents=rents[0]
        #print(parents[1])
            parents[1]=betterbasis
        children=mating(pop,parents,rcross,rmut,n_bits,n_mating)
        #print("breed")
            #print(parents[1])
        #print(children)
        for cc in range(len(children)):
            children[cc]= [int(ge) for ge in children[cc]]
        #print(children)
        if uselessgens==n_greed:
            uselessgens=0
            n_greed=n_greed+100
            children[1]=copy.deepcopy(betterbasis)
        removable=[int(k) for k in range(len(children))]
        #removable=[0,1]#worst2(weights)
        # newgen=[]
        # togo=[]
        # for pg in range(n_parents):
        #     rents,weights=selectparents(pop,fitness,n_mating)
        #     parents=rents[0]
        # #print(parents)
        #     children=mating(pop,parents,rcross,rmut,n_bits)
    
        #     #print(children)
        #     for cc in range(len(children)):
        #         children[cc]= [int(ge) for ge in children[cc]]
        #         #print(children)     
        #         removable=worst2(weights)
        #         togo.append(removable[0])
        #         togo.append(removable[1])
        #         newgen.append(children[0])
        #         newgen.append(children[1])
        #print(removable)
        for a,k in zip(removable,children):
            #print(a)
            #print(k)
            pop[a]=k
            fitness[a]=objectivesvd(k, sensitivity, D)
        pop = [x for _,x in sorted(zip(fitness,pop))]
        fitness = [x for x in sorted(fitness)]
    #print(pop)
    #for c in pop:
        #print(c)
    #print('here')
    fitnessnew = [objectivesvd(c, sensitivity, D) for c in pop]
    for p in range(n_pop):#[0,1,-1]:#range(n_pop):
        #print('here')
        if fitnessnew[p] > best_eval:
            #print('here2')
            best, best_eval = pop[p], fitness[p]
            elapsed_time = datetime.now()-start_time
            print("gen:%d,  duration: %s, new best: %s) = %E" % (gen, elapsed_time, pop[p], fitnessnew[p]))
    #fittestIndex = np.where(fitnessnew == np.max(fitnessnew))
    #print(fittestIndex)
    #Extracting index of fittest chromosome
    #fittestIndex = fittestIndex[0][0]
    # Getting Best chromosome
    #fittestInd = pop[fittestIndex]
    bestbasis=[]
    #print(best)
    for b in best:
        bestbasis.append(charliemeasurements[b])
    #bestFitness = fitness[fittestIndex]
    bestFitness= best_eval
    # print("Best:")
    # print(best)
    print("\nBest Individual:")
    print(bestbasis)
    print("\nBest Individual's Fitness:")
    print(bestFitness)
    end_time = datetime.now()
    print('Duration: {}'.format(end_time - start_time))
    return [bestFitness,bestbasis]
    #return [best_eval,format(end_time - start_time),best]

def genguide(chrom,charliemeasurements,d,n_bits,greedlim): #new_measurements
    maxvec=[]
    outchrom=[]
    for gene in chrom:
        #maxvec.append(gene)
        maxvec.append(copy.deepcopy(charliemeasurements[gene]))
    #print('check1')
    #checks for maximized basis by comparing if a new addition of one of the other measurements has a better perp component than the one added
    popindex=random.randint(0,n_bits-2)
    added=copy.deepcopy(maxvec[popindex])
    redos=0
    checking=copy.deepcopy(maxvec)
    checking.pop(maxvec.index(added))
    # for each in added:
    #     checking.pop(maxvec.index(each))
    #print('check2')
    new=copy.deepcopy(maxvec)
    checked=[]
    checked.append(copy.deepcopy(added))
    #print('check3')
    for each in checking:
        testbasis=copy.deepcopy(new)
        testbasis.pop(testbasis.index(each))
        mxvol,mxmes=mxvolchecklim(testbasis, charliemeasurements,greedlim)
        #print('check4')
        if each==mxmes:
            #print('NO CHANGE')
            checked.append(copy.deepcopy(each))
        else:
            redos=redos+1
            
            #print('NEW MEAS')
            for c in checked:
                checking.append(copy.deepcopy(c))
            #checking.append(copy.deepcopy(each))
            new.pop(new.index(each))
            new.append(copy.deepcopy(mxmes))
            checked=[]
            checked.append(copy.deepcopy(mxmes))    
            #redos=redos+1
            if redos>=d: 
                for gene in new:
                    outchrom.append(copy.deepcopy(charliemeasurements.index(gene)))
                return outchrom
    for gene in new:
        outchrom.append(copy.deepcopy(charliemeasurements.index(gene)))
    return outchrom

#Fix for larger values of C and D
def objectivesvd(chrom, sensitivity, D):
    sensmx=[]
    #print('newit')
    for gene in chrom:
        #print(gene)
        sensmx.append(1*copy.deepcopy(sensitivity[gene]))
    svs=svdvals(sensmx)
    #print(svs)
    if len(svs)<len(chrom):
        V=0
    if np.any(svs<1e-17):
        V=0
    else:
        V=np.prod(svs**(1/D))
    #print('Vol')
    #print(V)
    return V
def volume_checksvd(basis):
    sensmx=[]
    for each in basis:
        sensmx.append(1*copy.deepcopy(sensitivity[charliemeasurements.index(each)]))
    svs=svdvals(sensmx)
    if len(svs)<len(basis):
        V=0
    if np.any(svs<1e-17):
        V=0
    else:
        #V=np.prod(svs)
        V=np.prod(svs**(1/D))
    #polyvol=np.prod(svs)
    return V

#More fix for large #s
def mxvolchecklim(trimmedbasis, allmeas, fract):
    #start_time=datetime.now()
    #of the options available, picks the measurement that maximizes the volume 
    measurements = copy.deepcopy(trimmedbasis)
    polyvol, muprimes, muprimesmag = volume_check2(measurements)
    polyvol=volume_checksvd(measurements)
    mxvol=0
    totalmeas=random.sample(allmeas,int(np.floor(fract*len(allmeas))))
    trialbs=copy.deepcopy(trimmedbasis)
    for option in totalmeas:
        if (option not in trimmedbasis):
            thismes = copy.deepcopy(option)
            trialbs.append(thismes)
            #vol=volume_check(trialbs)
            d = len(trimmedbasis) 
            mu = sensitivity[charliemeasurements.index(thismes)]
            mupar = np.zeros(sensitivity.shape[1],)
            for val in range(d):
                mupar += (np.dot(muprimes[val], mu)/(muprimesmag[val]**2)) * muprimes[val] 
            muprime = mu - mupar
            muprimemag = np.linalg.norm(muprime)
            muprimemag = muprimemag**(1/D)
            vol=muprimemag*polyvol
            if vol>mxvol:
                mxvol=copy.deepcopy(vol)
                mxmes=copy.deepcopy(option)
    end_time = datetime.now()
    #print('Duration: {}'.format(end_time - start_time))
    return mxvol, mxmes     
```
